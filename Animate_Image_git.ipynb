```python
from pathlib import Path
base_frames_dir = Path("/content/frames")

#Collect all frame paths from the first three videos for training and the fourth vid for validation
train_dirs = []
for i in range(1, 4):
    video_frames_dir = base_frames_dir / f"video_{i}"
    train_dirs.extend(list(video_frames_dir.glob('*.png')))

val_dirs = list((base_frames_dir / "video_4").glob('*.png'))
```

```python
import torch
from torchvision.io import read_image
from torchvision.transforms import Compose, Resize, Normalize
from torch.utils.data import DataLoader, Dataset

transforms = Compose([
    Resize((128, 128)),  # Resize images to fit the model input
    Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Custom dataset to load frames based on given paths
class StrawberryDataset(Dataset):
    def __init__(self, frame_paths, transforms=None):
        self.frame_paths = frame_paths
        self.transforms = transforms

    def __len__(self):
        return len(self.frame_paths)

    def __getitem__(self, idx):
        frame_path = self.frame_paths[idx]
        frame = read_image(str(frame_path)).float() / 255
        if self.transforms:
            frame = self.transforms(frame)
        return frame

#dataset instances
train_dataset = StrawberryDataset(train_dirs, transforms=transforms)
val_dataset = StrawberryDataset(val_dirs, transforms=transforms)

#Creating DataLoaders
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False)
```

```python
import torch
from torch.optim import AdamW
from tqdm import tqdm
import torch.nn.functional as F
from torchvision.transforms.functional import to_tensor, resize
from torchvision.transforms import Resize

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

#optimizer
optimizer = AdamW(adapter.parameters(), lr=1e-5)

# Resize function
resize_transform = Resize((128, 128))

# Training loop
num_epochs = 2
for epoch in range(num_epochs):
    epoch_loss = 0
    print(f"Epoch {epoch+1}/{num_epochs}")

    #Iterate over DataLoader
    for frames in tqdm(train_loader, desc="Training", leave=False):
        frames = frames.to(device)
        optimizer.zero_grad()

        try:
            outputs = pipeline(prompt=["strawberry rotting"] * len(frames), image=frames)
            #generated frames to tensors
            generated_frames = torch.stack([to_tensor(frame).to(device) for frame in outputs.frames[0]])

            generated_frames = resize_transform(generated_frames)  # Resizing to match ground truth
            generated_frames.requires_grad_(True)

            if generated_frames.shape[0] != frames.shape[0]:
                frames = frames.expand_as(generated_frames)

            #loss calculation
            loss = F.mse_loss(generated_frames, frames)

            loss.backward()
            optimizer.step()

            batch_loss = loss.item()
            epoch_loss += batch_loss

            print(f"Batch Loss: {batch_loss:.4f}")
        except Exception as e:
            print(f"Error processing batch: {str(e)}")

    # Compute loss
    avg_epoch_loss = epoch_loss / len(train_loader)
    print(f"Average Epoch Loss: {avg_epoch_loss:.4f}")
```

```python
import os
from PIL import Image
from torchvision.transforms import Compose, Resize, Normalize, ToTensor
import matplotlib.pyplot as plt
from diffusers import AnimateDiffPipeline, MotionAdapter
import torch

os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'
device = "cuda" if torch.cuda.is_available() else "cpu"

#Load the motion adapter with the fine-tuned version
adapter = MotionAdapter.from_pretrained("guoyww/animatediff-motion-adapter-v1-5-2", torch_dtype=torch.float32)
fine_tuned_adapter_path = '/content/fine_tuned_adapter.pth'
adapter.load_state_dict(torch.load(fine_tuned_adapter_path))
adapter = adapter.to(device)

#loading the AnimateDiffPipeline with the fine-tuned adapter
pipeline = AnimateDiffPipeline.from_pretrained("emilianJR/epiCRealism", motion_adapter=adapter, torch_dtype=torch.float32)
pipeline = pipeline.to(device)

# Load and transform the input image
input_image_path = '/content/strawberry_image.png'
input_image = Image.open(input_image_path).convert("RGB")

#Resize and normalize the images for the model
transform = Compose([
    Resize((128, 128)),
    ToTensor(),
    Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
input_image = transform(input_image).unsqueeze(0).to(device)

# video generation parameters
output = pipeline(
    prompt="A strawberry rotting, high resolution",
    image=input_image,
    negative_prompt="bad quality, worse quality, low resolution",
    num_frames=16,
    guidance_scale=7.5,
    num_inference_steps=50,
    generator=torch.Generator(device).manual_seed(49),
)

frames = output.frames[0]
#Display the generated frames
plt.figure(figsize=(15, 10))
for i, frame in enumerate(frames):
    plt.subplot(1, len(frames), i + 1)
    plt.imshow(frame)
    plt.title(f"Frame {i+1}")
    plt.axis('off')
plt.show()
```
